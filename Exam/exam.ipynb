`[0.2]` Create a new column `whoWon` which contains `"red"` or `"blue"` based on the `blueWins` column.

````python
lol_data['whoWon'] = 0
lol_data['whoWon'] = np.where((lol_data['blueWins'] == 1),
                             'blue',
                              lol_data['whoWon']
                             )

lol_data['whoWon'] = np.where((lol_data['blueWins'] == 0),
                             'red',
                              lol_data['whoWon']
                             )
````

````python
lol_data['firstKill'] = 0
lol_data['firstKill'] = np.where((lol_data['blueFirstBlood'] == 1),
                             'blue',
                              lol_data['firstKill']
                             )

lol_data['firstKill'] = np.where((lol_data['redFirstBlood'] == 1),
                             'red',
                              lol_data['firstKill']
                             )

lol_data
````
